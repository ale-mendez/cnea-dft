```python
def BM(x, E0, B0, V0, B0p):
    return E0+(9/16)*V0*B0*(B0p*((V0/x)**(2/3)-1)**3+((V0/x)**(2/3)-1)**2*(6-4*(V0/x)**(2/3)))
```

```python
VBCC_GGA=BCC_GGA_mag["Volumen"]
EBCC_GGA=BCC_GGA_mag["Energía total"]
param_BCCGGA, cov_BCCGGA = optimize.curve_fit(BM, VBCC_GGA, EBCC_GGA,p0=[1,1,6,1])
```

```python
print("Parámetros de BCC GGA magnético:")
print("E0 = {:6.4f}".format(param_BCCGGA[0]))
print("B0 = {:6.4f}".format(param_BCCGGA[1]))
print("V0 = {:6.4f}".format(param_BCCGGA[2]))
print("B0' = {:6.4f}".format(param_BCCGGA[3]))
```

```python
VBCC_LDA=BCC_LDA_mag["Volumen"]
EBCC_LDA=BCC_LDA_mag["Energía total"]
param_BCCLDA, cov_BCCLDA = optimize.curve_fit(BM, VBCC_LDA, EBCC_LDA,p0=[1,1,6,1])
```

```python
print("Parámetros de BCC LDA magnético:")
print("E0 = {:6.4f}".format(param_BCCLDA[0]))
print("B0 = {:6.4f}".format(param_BCCLDA[1]))
print("V0 = {:6.4f}".format(param_BCCLDA[2]))
print("B0' = {:6.4f}".format(param_BCCLDA[3]))
```

```python
aGGA=(4*param_BCCGGA[2])**(1/3)
aLDA=(4*param_BCCLDA[2])**(1/3)
```

```python
print("a_GGA=",aGGA)
print("a_LDA=",aLDA)
```

```python
npts=100000
xfit=np.linspace(2,15,npts)
BCC_GGA_fit=BM(xfit,param_BCCGGA[0],param_BCCGGA[1],param_BCCGGA[2],param_BCCGGA[3])
BCC_LDA_fit=BM(xfit,param_BCCLDA[0],param_BCCLDA[1],param_BCCLDA[2],param_BCCLDA[3])
```

```python
fig=plt.figure(figsize=(10,8))
ax=plt.subplot(111)
ax.set_title('Fe con estructura BCC (magnético)',pad=20)
plt.plot(BCC_GGA_mag['Volumen'],BCC_GGA_mag['Energía total'],'ko',linestyle='none',linewidth=1,label='')
plt.plot(BCC_LDA_mag['Volumen'],BCC_LDA_mag['Energía total'],'ko',linestyle='none',linewidth=1,label='')
plt.plot(xfit,BCC_GGA_fit,'k',linestyle='solid',linewidth=1,label='GGA fitted')
plt.plot(xfit,BCC_LDA_fit,'k',linestyle='dashed',linewidth=1,label='LDA fitted')
ax.axvline(x=Vcelda_exp,c='gray',linestyle='dotted',linewidth=2)

ax.legend(bbox_to_anchor=(0.95,0.1), loc='lower right', borderaxespad=0.,
          ncol=1, handlelength=0.95,handletextpad=0.2,columnspacing=0.9)
ax.xaxis.set_major_locator(mticker.MultipleLocator(2));
ax.xaxis.set_minor_locator(mticker.MultipleLocator(0.5));
ax.yaxis.set_major_locator(mticker.MultipleLocator(0.5));
ax.yaxis.set_minor_locator(mticker.MultipleLocator(0.1));
ax.tick_params(direction='in',which='major',length=8,bottom=True, top=True, left=True, right=True);
ax.tick_params(direction='in',which='minor',length=5,bottom=True, top=True, left=True, right=True);
plt.ylabel('Energía total (eV)',labelpad=10, fontsize=26)
plt.xlabel(r'Volumen (\AA$^3$)',labelpad=10, fontsize=26)
ax.text(6.2,-4.9,r"$V_{exp}=5.88$\AA$^3$",fontsize=22)
plt.ylim(-5.1,-2.9)
plt.xlim(1.5,14)
plt.tight_layout()
plt.savefig("punto7.eps")
plt.show()
```
